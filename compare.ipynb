```python
from optimizer2 import solve
import pandas as pd
import numpy as np
from BRL import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tests = pd.read_excel('./data/data.xlsx', index_col=0).iloc[:,1:].values
tests.shape
```

lambda = 0.1

```python
similarities = []
for test in tests:
    solution = solve(e_T=test, w2=589, w3=617, w5=978, w6=211, w8=223, w9=793)
    x = np.zeros(66)
    if bool(solution):
        for i in solution:
            x[i] = solution[i]
        e = p_matrix@x
        similarity = np.dot(e, tests[i])/(np.linalg.norm(e)*np.linalg.norm(tests[i]))
        similarities.append(similarity)
```

```python
print(similarities)
```

```python
sns.kdeplot(similarities)
plt.show()
```

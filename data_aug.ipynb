```python
temp = np.random.uniform(0.45, 0.6, 120)
input_volume['R104'] = (input_volume['R106']*temp).astype(int)
output_volume['R104'] = (output_volume['R106']*temp).astype(int)
```

```python
def data_aug(df, lb, ub, oneborn, aug):
    temp = np.random.uniform(lb, ub, 120)
    df[aug] = (df[oneborn]*temp).astype(int)
    return df
```
